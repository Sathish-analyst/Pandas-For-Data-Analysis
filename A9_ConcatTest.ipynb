```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
india_weather_df=pd.DataFrame(
    {
    "city":['chennai','mumbai','bangalor'],
    "temp":[35,45,55],
    "wind":[60,70,80]
    }
)

india_weather_df
```

```python
us_weather_df=pd.DataFrame(
    {
    "city":['Newyork','chicago','Dallas'],
    "temp":[30,40,50],
    "wind":[61,71,81]
    }
)

us_weather_df
```

```python
concat_df=pd.concat([india_weather_df,us_weather_df])
concat_df
```

```python
concat_df=pd.concat([india_weather_df,us_weather_df],ignore_index=True)
concat_df
```

```python
concat_df=pd.concat([india_weather_df,us_weather_df],keys=['india','us'])
concat_df
```

```python
india_weather_df=pd.DataFrame(
    {
    "city":['chennai','mumbai','bangalor'],
    "temp":[35,45,55]
    }
)
india_weather_df
```
